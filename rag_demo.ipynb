```python
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from groq import Groq
```

```python
# 2. Loading the pdf
loader=PyPDFLoader(r"C:\Users\HP\Desktop\RAG Project\research_pdf.pdf")
documents=loader.load()
```

```python
# 3. split the document
text_splitter=RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks=text_splitter.split_documents(documents)
```

```python
# 4. Creating Embeddings
embeddings=HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
# 5. Building the Vector databases
vector_store=FAISS.from_documents(
    chunks,
    embeddings
)
```

```python
# 6. Initializing the LLM
client = Groq(
    api_key=os.getenv("GROQ_API_KEY")
)
```

```python
# 7. Interavtive Question Loop
while True:
  question=input("Ask the Question : ")
  if question.lower()=="exit":
    break
  # 8. semantic search
  docs=vector_store.similarity_search(
      question,
      k=3
  )
  # 9. Creating Context
  context="\n\n".join(doc.page_content for doc in docs)

  # 10. prompt engineering
  prompt=f"""
  Use the following context to answer the question.

  Context:
  {context}

  Question:
  {question}

  Answer clearly
  """

  # 11. Sending request to LLM
  response=client.chat.completions.create(
      model="llama-3.1-8b-instant",
      messages=[
          {"role":"user","content":prompt}
      ]
  )

  # 12. Extracting the answer
  answer=response.choices[0].message.content

  # 13. print answer
  print("\n AI Answer : \n ",answer)
```
